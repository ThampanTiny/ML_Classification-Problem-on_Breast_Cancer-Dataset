###LOGISTIC REGRESSION

Logistic regression is a supervised machine learning algorithm used for classification tasks where the goal is to predict the probability that an instance belongs to a given class or not. For example, we have two classes Class 0 and Class 1 if the value of the logistic function for an input is greater than 0.5 (threshold value) then it belongs to Class 1 otherwise it belongs to Class 0. Its referred to as regression because it is the extension of linear regression but is mainly used for classification problems. It  is often chosen for medical data because it excels at predicting binary outcomes (like "disease present" or "disease absent") based on various patient factors, making it ideal for analyzing situations where the goal is to classify patients into two distinct categories based on their medical characteristics. 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_pred
```

### DECISION TREE CLASSIFIER

A decision tree classifier is a machine learning algorithm that uses a tree-like structure to classify data based on its features. It's a supervised learning algorithm that can be used for classification and regression modeling. It is  chosen for medical data analysis because it can handle a mix of categorical and continuous variables, is easy to interpret, and can effectively identify the most important factors contributing to a diagnosis or treatment outcome, making it valuable for understanding complex medical decision-making processes

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = DecisionTreeClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print('\033[1m###########  DECISION TREE CLASSIFIER  ###########\033[0m')

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.3f}")

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="coolwarm")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

performance['DECISION TREE'] = {"Accuracy": round(accuracy,3)}

nvd = pd.DataFrame(classification_report(y_test, y_pred,output_dict=True))
performance_df['F1-score']['DECISION TREE']=nvd['accuracy']['f1-score']
performance_df
```

### RANDOM FOREST CLASSIFIER

Random Forest is a supervised machine learning algorithm made up of decision trees. It  is often chosen for analyzing medical data because of its ability to handle high-dimensional datasets, identify important features within complex medical information, and provide robust predictions even with noisy data, making it suitable for tasks like disease diagnosis and prognosis prediction; essentially, it offers a good balance between accuracy and interpretability in the medical field.

```python
from sklearn.ensemble  import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print('\033[1m###########  RANDOM FOREST CLASSIFIER  ###########\033[0m')

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.3f}")

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="coolwarm")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

performance['RANDOM FOREST'] = {"Accuracy": round(accuracy,3)}

nvd = pd.DataFrame(classification_report(y_test, y_pred,output_dict=True))
performance_df['F1-score']['RANDOM FOREST']=nvd['accuracy']['f1-score']
performance_df
```

### SUPPORT VECTOR MACHINE (SVM)

A support vector machine (SVM) is a supervised machine learning algorithm that classifies data by finding an optimal line or hyperplane that maximizes the distance between each class in an N-dimensional space. SVMs are also less sensitive to overfitting and easier to interpret. So it performs better compared to Linear Classification and decision Tree.

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print('\033[1m###########  SUPPORT VECTOR MACHINE CLASSIFIER  ###########\033[0m')

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.3f}")

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="coolwarm")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

performance['SVM'] = {"Accuracy": round(accuracy,3)}

nvd = pd.DataFrame(classification_report(y_test, y_pred,output_dict=True))
performance_df['F1-score']['SVM']=nvd['accuracy']['f1-score']
performance_df
```

### K-NEAREST NEIGHBOUR (KNN)

K-Nearest Neighbors (KNN) algorithm is a machine learning tool that classifies data by comparing it to similar nearby data points. It's used for pattern recognition tasks, such as classifying objects into different categories. It can effectively classify new patient cases by comparing them to similar historical cases in the training data based on their features. 

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('\033[1m###########  K-NEAREST NEIGHBOUR CLASSIFIER  ###########\033[0m')

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.3f}")

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="coolwarm")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

performance['KNN'] = {"Accuracy": round(accuracy,3)}

nvd = pd.DataFrame(classification_report(y_test, y_pred,output_dict=True))
performance_df['F1-score']['KNN']=nvd['accuracy']['f1-score']
performance_df
```
